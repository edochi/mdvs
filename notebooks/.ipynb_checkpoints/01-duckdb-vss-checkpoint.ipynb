# Spike 01: DuckDB + Vector Search

**Goal:** Validate that `duckdb-rs` can:
1. Open an in-memory database
2. Create a table with `FLOAT[256]` column
3. Insert `Vec<f32>` into that column
4. Query with `array_cosine_distance()` (built-in, no extension needed)
5. JOIN across tables with vector search

**Note:** `array_cosine_distance()` is a built-in DuckDB function. The `vss` extension
is only needed for HNSW indexing (approximate NN for large datasets). Skipping it for now —
brute-force exact search is fast enough for typical vault sizes (<100k chunks).

```rust
:dep duckdb = { version = "*", features = ["bundled"] }
```

## 1. Open DB

```rust
use duckdb::{Connection, params};

let conn = Connection::open_in_memory().unwrap();
println!("DuckDB connection opened successfully");
```

## 2. Create tables with FLOAT[256] column

```rust
conn.execute_batch("
    CREATE TABLE mdfiles (
        filename VARCHAR PRIMARY KEY,
        title VARCHAR,
        content_hash VARCHAR
    );

    CREATE TABLE chunks (
        chunk_id VARCHAR PRIMARY KEY,
        filename VARCHAR NOT NULL REFERENCES mdfiles(filename),
        chunk_index INTEGER,
        heading VARCHAR,
        plain_text VARCHAR,
        embedding FLOAT[256],
        char_count INTEGER
    );
").unwrap();

println!("Tables created with FLOAT[256] column");
```

## 3. Insert Vec<f32> into FLOAT[256]

`Vec<f32>` doesn't implement `ToSql` in duckdb-rs, so `params!` won't work.
Testing SQL literal approach: format the vector as `[0.0, 0.001, ...]::FLOAT[256]`.

```rust
// Helper: format Vec<f32> as a DuckDB array literal
fn embedding_to_sql(embedding: &[f32], dim: usize) -> String {
    format!(
        "[{}]::FLOAT[{}]",
        embedding.iter().map(|f| f.to_string()).collect::<Vec<_>>().join(","),
        dim
    )
}

// Insert a file
conn.execute(
    "INSERT INTO mdfiles (filename, title, content_hash) VALUES (?, ?, ?)",
    params!["test.md", "Test Note", "abc123"],
).unwrap();

// Create a fake embedding (256 floats)
let embedding: Vec<f32> = (0..256).map(|i| (i as f32) / 256.0).collect();

// Insert chunk with embedding as SQL literal
let sql = format!(
    "INSERT INTO chunks (chunk_id, filename, chunk_index, heading, plain_text, embedding, char_count) \
     VALUES ('test.md#0', 'test.md', 0, 'Test Heading', 'test plain text', {}, 15)",
    embedding_to_sql(&embedding, 256)
);

conn.execute(&sql, []).unwrap();
println!("Inserted chunk with embedding via SQL literal");
```

## 4. Insert second row and test search

```rust
// Insert a second chunk so we have something to search against
let embedding2: Vec<f32> = (0..256).map(|i| 1.0 - (i as f32) / 256.0).collect();

conn.execute(
    "INSERT INTO mdfiles (filename, title, content_hash) VALUES (?, ?, ?)",
    params!["test2.md", "Second Note", "def456"],
).unwrap();

let sql2 = format!(
    "INSERT INTO chunks (chunk_id, filename, chunk_index, heading, plain_text, embedding, char_count) \
     VALUES ('test2.md#0', 'test2.md', 0, 'Other Topic', 'different content here', {}, 21)",
    embedding_to_sql(&embedding2, 256)
);

conn.execute(&sql2, []).unwrap();
println!("Second row inserted");
```

## 5. Query with array_cosine_distance

```rust
// Create a query embedding (similar to embedding 1)
let query_vec: Vec<f32> = (0..256).map(|i| (i as f32) / 256.0 + 0.01).collect();
let query_literal = embedding_to_sql(&query_vec, 256);

{
    let query_sql = format!(
        "SELECT c.chunk_id, c.filename, c.heading, LEFT(c.plain_text, 50) as snippet, \
         array_cosine_distance(c.embedding, {}) AS distance \
         FROM chunks c \
         ORDER BY distance \
         LIMIT 10",
        query_literal
    );

    let mut stmt = conn.prepare(&query_sql).unwrap();
    let results = stmt.query_map([], |row| {
        Ok((
            row.get::<_, String>(0)?,
            row.get::<_, String>(1)?,
            row.get::<_, Option<String>>(2)?,
            row.get::<_, String>(3)?,
            row.get::<_, f64>(4)?,
        ))
    }).unwrap();

    for result in results {
        let (chunk_id, filename, heading, snippet, distance) = result.unwrap();
        println!("{:.4}  {} § {:?} — {}", distance, filename, heading, snippet);
    }
}
```

## 6. JOIN query (note-level ranking)

```rust
{
    let join_sql = format!(
        "WITH ranked_chunks AS ( \
             SELECT c.filename, c.heading, LEFT(c.plain_text, 120) AS snippet, \
                    array_cosine_distance(c.embedding, {}) AS distance \
             FROM chunks c \
         ) \
         SELECT m.filename, m.title, \
                MIN(rc.distance) AS distance, \
                FIRST(rc.heading ORDER BY rc.distance) AS best_heading, \
                FIRST(rc.snippet ORDER BY rc.distance) AS snippet \
         FROM ranked_chunks rc \
         JOIN mdfiles m ON rc.filename = m.filename \
         GROUP BY m.filename, m.title \
         ORDER BY distance \
         LIMIT 10",
        query_literal
    );

    let mut stmt = conn.prepare(&join_sql).unwrap();
    let results = stmt.query_map([], |row| {
        Ok((
            row.get::<_, String>(0)?,
            row.get::<_, Option<String>>(1)?,
            row.get::<_, f64>(2)?,
            row.get::<_, Option<String>>(3)?,
            row.get::<_, String>(4)?,
        ))
    }).unwrap();

    println!("--- Note-Level Results ---");
    for result in results {
        let (filename, title, distance, heading, snippet) = result.unwrap();
        println!("{:.4}  {} ({:?}) § {:?}", distance, filename, title, heading);
        println!("        {}", snippet);
    }
}
```

## Findings

Record what worked and what didn't:

- [x] `params!` with `Vec<f32>` for FLOAT[N] — **doesn't work**, `Vec<f32>: ToSql` not implemented
- [x] SQL literal approach (`[0.0,0.1,...]::FLOAT[256]`) — workaround for inserts and queries
- [ ] `array_cosine_distance` — works as built-in?
- [ ] JOIN with cosine distance — note-level ranking query works?
