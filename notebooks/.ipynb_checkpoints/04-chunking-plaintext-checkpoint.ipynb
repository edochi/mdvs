# Spike 04: text-splitter + pulldown-cmark

**Goal:** Validate that:
1. `text-splitter` MarkdownSplitter produces sensible chunks
2. Chunks respect the max size parameter
3. `pulldown-cmark` can strip markdown to clean plain text
4. We can extract headings from chunks
5. Different note shapes (short, long, with/without headings) work

```rust
:dep text-splitter = { version = "*", features = ["markdown"] }
:dep pulldown-cmark = "*"
```

## 1. Basic chunking with MarkdownSplitter

```rust
use text_splitter::MarkdownSplitter;

let splitter = MarkdownSplitter::new(1000); // max 1000 chars per chunk

{
    let markdown = std::fs::read_to_string("../tests/fixtures/long-note.md").unwrap();

    // Remove frontmatter manually for now (gray_matter would do this)
    let body = if markdown.starts_with("---") {
        if let Some(end) = markdown[3..].find("---") {
            &markdown[end + 6..]
        } else {
            &markdown
        }
    } else {
        &markdown
    };

    let chunks: Vec<&str> = splitter.chunks(body).collect();

    println!("Total body length: {} chars", body.len());
    println!("Number of chunks: {}\n", chunks.len());

    for (i, chunk) in chunks.iter().enumerate() {
        println!("--- Chunk {} ({} chars) ---", i, chunk.len());
        let preview = if chunk.len() > 200 { &chunk[..200] } else { chunk };
        println!("{}\n", preview);
    }
}
```

## 2. Short note (should be a single chunk)

```rust
{
    let short = "A very short note that fits in a single chunk.";
    let chunks: Vec<&str> = splitter.chunks(short).collect();
    println!("Short note: {} chunks", chunks.len());
    for c in &chunks {
        println!("  [{} chars] {}", c.len(), c);
    }
}
```

## 3. Plain text extraction with pulldown-cmark

```rust
use pulldown_cmark::{Parser, Event, Tag};

fn extract_plain_text(markdown: &str) -> String {
    let parser = Parser::new(markdown);
    let mut text = String::new();
    for event in parser {
        if let Event::Text(t) = event {
            text.push_str(&t);
            text.push(' ');
        }
    }
    text.trim().to_string()
}

{
    let rich = std::fs::read_to_string("../tests/fixtures/rich-formatting.md").unwrap();
    let body = if rich.starts_with("---") {
        if let Some(end) = rich[3..].find("---") {
            rich[end + 6..].to_string()
        } else {
            rich
        }
    } else {
        rich
    };

    println!("=== Original markdown ===\n{}", &body[..500.min(body.len())]);
    println!("\n=== Extracted plain text ===\n{}", extract_plain_text(&body));
}
```

## 4. Heading extraction from chunks

```rust
use pulldown_cmark::TagEnd;

fn extract_heading(markdown: &str) -> Option<String> {
    let parser = Parser::new(markdown);
    let mut in_heading = false;
    let mut heading_text = String::new();

    for event in parser {
        match event {
            Event::Start(Tag::Heading { .. }) => {
                in_heading = true;
                heading_text.clear();
            }
            Event::End(TagEnd::Heading(_)) => {
                if in_heading && !heading_text.is_empty() {
                    return Some(heading_text.trim().to_string());
                }
                in_heading = false;
            }
            Event::Text(t) if in_heading => {
                heading_text.push_str(&t);
            }
            _ => {}
        }
    }
    None
}

{
    let markdown = std::fs::read_to_string("../tests/fixtures/long-note.md").unwrap();
    let body = if markdown.starts_with("---") {
        if let Some(end) = markdown[3..].find("---") {
            markdown[end + 6..].to_string()
        } else {
            markdown
        }
    } else {
        markdown
    };

    let chunks: Vec<&str> = splitter.chunks(&body).collect();
    for (i, chunk) in chunks.iter().enumerate() {
        let heading = extract_heading(chunk);
        println!("Chunk {}: heading = {:?}", i, heading);
    }
}
```

## 5. Different chunk sizes

```rust
{
    let markdown = std::fs::read_to_string("../tests/fixtures/long-note.md").unwrap();
    let body = if markdown.starts_with("---") {
        if let Some(end) = markdown[3..].find("---") {
            markdown[end + 6..].to_string()
        } else {
            markdown
        }
    } else {
        markdown
    };

    for max_size in [500, 1000, 2000] {
        let s = MarkdownSplitter::new(max_size);
        let chunks: Vec<&str> = s.chunks(&body).collect();
        let sizes: Vec<usize> = chunks.iter().map(|c| c.len()).collect();
        println!(
            "max_size={}: {} chunks, sizes={:?}",
            max_size, chunks.len(), sizes
        );
        let violations: Vec<_> = sizes.iter().filter(|&&s| s > max_size).collect();
        if violations.is_empty() {
            println!("  All chunks within limit");
        } else {
            println!("  VIOLATIONS: {:?}", violations);
        }
    }
}
```

## 6. Unicode: chars vs bytes

`MarkdownSplitter::new(max_size)` limits by **character count**, not byte count.
Rust's `str::len()` returns bytes. For ASCII they match, but for multi-byte Unicode
(CJK, emoji, accented chars) they diverge. Verify the splitter still respects the
character limit and that we measure correctly.

```rust
{
    // Mixed-script markdown: CJK, emoji, accented Latin
    let unicode_md = r#"# 分散システムの設計

分散システム（Distributed Systems）は、ネットワークで接続された複数のコンピュータが協調して動作するシステムです。

## コンセンサスアルゴリズム 🔧

Raftアルゴリズムは、リーダー選出とログ複製を通じてコンセンサスを達成します。各ノードは「フォロワー」「候補者」「リーダー」のいずれかの状態を持ちます。

### Paxosとの比較

Paxosは理論的には優れていますが、実装が複雑です。Raftは理解しやすさを重視して設計されました。

> 「シンプルさは究極の洗練である」— レオナルド・ダ・ヴィンチ 🎨

## Résumé des résultats

Les systèmes distribués nécessitent une coordination soigneuse. Voici les points clés:

- **Tolérance aux pannes** — le système continue de fonctionner même si certains nœuds tombent en panne
- **Cohérence** — tous les nœuds voient les mêmes données
- **Disponibilité** — le système répond toujours aux requêtes

### Émojis dans le contexte technique 🚀

Les émojis sont de plus en plus utilisés dans la documentation technique:
- ✅ Tests passés
- ❌ Échec
- ⚠️ Avertissement
- 🔄 En cours de synchronisation
"#;

    let max_chars = 200;
    let s = MarkdownSplitter::new(max_chars);
    let chunks: Vec<&str> = s.chunks(unicode_md).collect();

    println!("=== Unicode chunking (max_size={} chars) ===\n", max_chars);

    let mut any_char_violation = false;
    for (i, chunk) in chunks.iter().enumerate() {
        let byte_len = chunk.len();
        let char_len = chunk.chars().count();
        let ratio = byte_len as f64 / char_len as f64;
        let over = if char_len > max_chars { " ** CHAR VIOLATION **" } else { "" };
        if char_len > max_chars { any_char_violation = true; }

        println!(
            "Chunk {}: {} chars, {} bytes (ratio {:.2}x){}",
            i, char_len, byte_len, ratio, over
        );
        // Show first line as preview
        let first_line = chunk.lines().next().unwrap_or("");
        println!("  > {}\n", first_line);
    }

    println!("---");
    println!("Character limit respected: {}", !any_char_violation);

    // Show that .len() (bytes) would mislead us
    let byte_sizes: Vec<usize> = chunks.iter().map(|c| c.len()).collect();
    let char_sizes: Vec<usize> = chunks.iter().map(|c| c.chars().count()).collect();
    let byte_violations: Vec<_> = byte_sizes.iter().filter(|&&s| s > max_chars).collect();

    println!(
        "\nByte sizes:  {:?}\nChar sizes:  {:?}",
        byte_sizes, char_sizes
    );
    println!(
        "Chunks exceeding {} using .len() (bytes): {} — these are FALSE positives",
        max_chars, byte_violations.len()
    );
}
```

## Findings

- [x] `MarkdownSplitter::new(max_size)`, `.chunks(body)` returns iterator of `&str`
- [x] Respects max_size **strictly** — no violations at any size
- [x] Splits at heading boundaries first, then paragraphs, then sentences
- [x] Short notes → single chunk
- [x] Heading extraction via pulldown-cmark — works reliably, each chunk gets its heading
- [x] Plain text extraction via `Event::Text` — works for most content

**Plain text extraction issues to address in implementation:**
- `Event::Code` (inline code) is separate from `Event::Text` — need to include it
- `~~strikethrough~~` not parsed by default — need `Options::ENABLE_STRIKETHROUGH`
- `[[wikilinks]]` pass through as literal brackets (not standard markdown)
- Code block content (`Event::Text` inside `Tag::CodeBlock`) is included — may want to skip for embeddings
- Table pipe syntax passes through raw — need to extract cell text properly
